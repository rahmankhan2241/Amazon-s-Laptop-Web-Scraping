<h1 style = 'color: orange'><b>Scraping Amazon Laptop's Data</b> </h1>
<b> Aim: </b> Aim of this project is to scrap the <b>amazon's</b> website and get information of <b>laptop's</b> that are under <b>50K (Name, Price, Rating and Image Url)</b> </p>

```python
#Importing Necessary Libraries
import pandas as pd
from bs4 import BeautifulSoup
import requests

# Creating Dictiorney to store the values
specifications = {
        'name': [],
        'price': [],
        'Rating': [],
        'image_url': []
    }

# Looping over 20 pages
for k in range(21):

## Getting Url and getting the html format of the Web
    url = f'https://www.amazon.in/s?k=laptop+under+50000&page={k}'
    html_cont = requests.get(url).text
    soup = BeautifulSoup(html_cont, 'html.parser')

## Getting all the container containing the laptop's specification
    container = soup.find_all("div", class_ = "sg-col-inner")
## Looping over each container
    for i in container:
        price = i.find("span", class_= "a-price-whole")
        if price and int(price.text.strip().replace(',', "")) < 50000:
            for j in price:
                if int(j.text.strip().replace(',',"")) < 50000:
                    specifications['name'].append(i.find("span", class_= "a-size-medium a-color-base a-text-normal").text.strip())
                    specifications['price'].append(j.text.strip())
                    if i.find("span", class_= "a-icon-alt") != None:
                        specifications['Rating'].append(i.find("span", class_= "a-icon-alt").text.strip())
                    else:
                        specifications['Rating'].append("Null")
                    specifications['image_url'].append(i.find("img", class_= "s-image")["src"]) 
## Creating DataFrame of the information     
laptop_data = pd.DataFrame(specifications)

## Incase any duplicate value occur
laptop_data.drop_duplicates()

laptop_data
```
